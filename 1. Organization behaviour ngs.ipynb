```python
#Descriptive Statistics: Summarize the data to get a sense of how organizations are promoting cybersecurity awareness and training. 
#This includes calculating means, medians, and standard deviations for relevant questions.
import pandas as pd
```

```python
def create_frequency_mapping(df, columns):
    unique_values = set()
    
    for col in columns:
        print(f"{col} {df[col].unique()}")
        unique_values.update(df[col].unique())
        
    
    unique_values = sorted(list(unique_values))
    print(f"-------\n-------------\n {unique_values} ")
    mapping = {value: index for index, value in enumerate(unique_values)}
    
    return mapping
```

```python
file_path = "cleaneddata_NGS.csv"
df = pd.read_csv(file_path)
```

```python
# Relevant columns to convert
columns_to_convert = [
    #17----1
    'How often does your organization provide cybersecurity training specifically tailored to NGS operations',
    #19------2
    'Does your organization have clear and accessible cybersecurity policies specific to NGS operations',
    #20--------3
    'Have you experienced any cybersecurity incidents or breaches related to NGS data in the past',
    #21-----------------4
    'How often do you participate in cybersecurity training or awareness programs specifically tailored for NGS data security',
    #26-----------------------5
    'How relevant do you find the cybersecurity training content to your day-to-day responsibilities',
    #27----------------------------6
    'Since the last cybersecurity training session, have you encountered situations where you applied the knowledge learned',
    #28-------------------------------7
    'In your opinion, what are the biggest barriers to implementing effective cybersecurity training in your organization (Select all that apply)',
    #30---------------------------------8
    'Has your organization ever experienced a cybersecurity breach that affected NGS or any other sensitive data',
    #35--------------------------------------9
    'Has there been a cybersecurity incident within your organization such as hacking, password stealing etc.,  that led to changes or enhancements in the training program or policy change'
]
```

```python
# Create the frequency mapping
frequency_mapping = create_frequency_mapping(df, columns_to_convert)
```

```python
# Convert relevant columns using the mapping
for col in columns_to_convert:
    df[col] = df[col].map(frequency_mapping)
```

```python
# Calculate means, medians, and standard deviations
summary_stats = df[columns_to_convert].agg(['mean', 'median', 'std'])
```

```python
#ANOVA (Analysis of Variance):
#Determine if there are significant differences in cybersecurity promotion efforts across different groups, 
#such as departments or roles.
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
def create_frequency_mapping(df, columns):
    unique_values = set()
    for col in columns:
        unique_values.update(df[col].dropna().unique())
    
    unique_values = sorted(list(unique_values))
    mapping = {value: index for index, value in enumerate(unique_values)}
    
    return mapping
```

```python
file_path = "cleaneddata_NGS.csv"
df = pd.read_csv(file_path)
```

```python
# Relevant columns to convert
columns_to_analyze = [
   #17
    'How often does your organization provide cybersecurity training specifically tailored to NGS operations',
    #19
    'Does your organization have clear and accessible cybersecurity policies specific to NGS operations',
    #20
    'Have you experienced any cybersecurity incidents or breaches related to NGS data in the past',
    #21
    'How often do you participate in cybersecurity training or awareness programs specifically tailored for NGS data security',
    #26
    'How relevant do you find the cybersecurity training content to your day-to-day responsibilities',
    #27
    'Since the last cybersecurity training session, have you encountered situations where you applied the knowledge learned',
    #28
    'In your opinion, what are the biggest barriers to implementing effective cybersecurity training in your organization (Select all that apply)',
    #30
    'Has your organization ever experienced a cybersecurity breach that affected NGS or any other sensitive data',
    #35
    'Has there been a cybersecurity incident within your organization such as hacking, password stealing etc.,  that led to changes or enhancements in the training program or policy change'
]
```

```python
# Rename columns to be more formula-friendly
df.rename(columns={
#17
    'How often does your organization provide cybersecurity training specifically tailored to NGS operations':'org_training',
    #19
    'Does your organization have clear and accessible cybersecurity policies specific to NGS operations': 'cs_policies',
    #20
    'Have you experienced any cybersecurity incidents or breaches related to NGS data in the past': 'cs_incidents',
    #21
    'How often do you participate in cybersecurity training or awareness programs specifically tailored for NGS data security':'cs_training',
    #26
    'How relevant do you find the cybersecurity training content to your day-to-day responsibilities':'cs_training_inlife',
    #27
    'Since the last cybersecurity training session, have you encountered situations where you applied the knowledge learned': 'cs_training_knowledge',
    #28
    'In your opinion, what are the biggest barriers to implementing effective cybersecurity training in your organization (Select all that apply)': 'cs_barriers',
    #30
    'Has your organization ever experienced a cybersecurity breach that affected NGS or any other sensitive data':'cs_breach_experienced',
    #35
    'Has there been a cybersecurity incident within your organization such as hacking, password stealing etc.,  that led to changes or enhancements in the training program or policy change': 'training_program_change'

}, inplace=True)
```

```python
# Update columns_to_analyze with the renamed columns
columns_to_analyze = [
    #17
  'org_training',
    #19
   'cs_policies',
    #20
    'cs_incidents',
    #21
   'cs_training',
    #26
     'cs_training_inlife',
    #27
  'cs_training_knowledge',
    #28
      'cs_barriers',
    #30
    'cs_breach_experienced',
    #35
  'training_program_change'
   
]
```

```python
# Create the frequency mapping
frequency_mapping = create_frequency_mapping(df, columns_to_analyze)
```

```python
# Convert relevant columns using the mapping
for col in columns_to_analyze:
    df[col] = df[col].map(frequency_mapping)
```

```python
# Drop rows with missing values in the relevant columns
df_cleaned = df.dropna(subset=columns_to_analyze)
```

```python
# Check the count of roles after dropping rows with missing values
role_counts_cleaned = df_cleaned['org_training'].value_counts()
print(role_counts_cleaned)
```

```python
# Perform ANOVA for each column
anova_results = {}

for col in columns_to_analyze:
    model = ols(f'{col} ~ C(role)', data=df_cleaned).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    anova_results[col] = anova_table
```

```python
# Display the ANOVA results
anova_results
```
