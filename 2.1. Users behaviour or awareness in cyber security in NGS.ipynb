```python
# Function to map categorical responses to numerical values
def create_frequency_mapping(df, columns):
    unique_values = set()
    for col in columns:
        unique_values.update(df[col].dropna().unique())
    
    unique_values = sorted(list(unique_values))
    mapping = {value: index for index, value in enumerate(unique_values)}
    
    return mapping
```

```python
df.columns
```

```python
# Create the frequency mapping

frequency_mapping = create_frequency_mapping(df, columns_to_analyze)
```

```python
# Convert relevant columns using the mapping
#for col in columns_to_analyze[:-1]:
    #df[col] = df[col].map(frequency_mapping)

for col in columns_to_analyze:
    df[col] = df[col].map(frequency_mapping)
```

```python
# Calculate descriptive statistics: counts, means, medians, and standard deviations
descriptive_stats = df[columns_to_analyze].agg(['count', 'mean', 'median', 'std'])
pd.set_option('display.max_columns', None)
# Display the descriptive statistics
print(descriptive_stats)
```
